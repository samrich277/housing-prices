```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://statistics.byu.edu/directory"
```

```python
r = requests.get(url)
print(r.status_code)
soup = BeautifulSoup(r.text)
container = soup.find('main', {'class': 'Page-main'})
print(container)
items = soup.find_all('div', {'class':'ListImageOnTopCircular-items-item'})
print(items)


```

```python
soup = BeautifulSoup(r.text)
container = soup.find('main', {'class': 'Page-main'})
items = soup.find_all('div', {'class':'ListImageOnTopCircular-items-item'})


h3 = items[0].find('h3')
name = h3.text.strip()
link = h3.find('a')['href']
print(name)
print(link)




```

```python
names = []
links = []
for item in items:
    h3 = item.find('h3')
    name = h3.text.strip()
    link = h3.find('a')['href']
    names.append(name)
    links.append(link)

print(names)
print(links)


    
```

```python
#MAKE A DATAFRAME
df = pd.DataFrame({'name': names, 'link': links})
df.head()
```
